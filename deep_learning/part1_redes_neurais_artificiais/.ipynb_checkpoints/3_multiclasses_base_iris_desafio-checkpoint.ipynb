```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import utils as np_utils
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

## Convertendo os dados categoricos nominais em ordinais (apenas no alvo y)

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y
```

```python
# Precisa-se fazer a conversão abaixo para que a estrutura da rede neural consiga entender as saidas de cada classificacao
y = np_utils.to_categorical(y)
y
```

```python
y.shape
```

## Divisão de base de treino e teste

```python
X_treinamento , X_teste, y_treinamento , y_teste = train_test_split(X,y,test_size=0.25)
```

```python
X_treinamento.shape , X_teste.shape, y_treinamento.shape , y_teste.shape
```

## Estrutura Rede Neural

```python
# Para calcular a quantidade de neurônios na cadama oculta

# numero de entrada (previsores) + numeros de saidas (alvo) / 2

(4+3)/2
```

```python
rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape=(4,)),
    tf.keras.layers.Dense(units=4,activation='relu'),
    tf.keras.layers.Dropout(rate = 0.2),
    tf.keras.layers.Dense(units=4,activation='relu'),
    tf.keras.layers.Dropout(rate = 0.2),
    tf.keras.layers.Dense(units=3,activation='softmax'),
])
```

```python
rede_neural.summary()
```

```python
rede_neural.compile(optimizer = 'adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
rede_neural.fit(X_treinamento,y_treinamento, batch_size=10, epochs=2000)
```

## Avaliando desempenho Rede Neural

```python
rede_neural.evaluate(X_teste,y_teste)
```

```python
previsoes = rede_neural.predict(X_teste)
print(previsoes)
```

```python
previsoes = previsoes > 0.5
print(previsoes)
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import utils as np_utils
from tensorflow.keras import backend as k
from tensorflow.keras.models import Sequential
from sklearn.model_selection import cross_val_score
```

```python
def criar_rede():
    k.clear_session()
    rede_neural = Sequential([
        tf.keras.layers.InputLayer(shape=(4,)),
        tf.keras.layers.Dense(units=5,activation='relu'),
        tf.keras.layers.Dropout(rate = 0.2),
        tf.keras.layers.Dense(units=5,activation='relu'),
        tf.keras.layers.Dropout(rate = 0.2),
        tf.keras.layers.Dense(units=3,activation='softmax'),
    ])
    rede_neural.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['categorical_accuracy'])
    return rede_neural
```

```python
rede_neural = KerasClassifier(model = criar_rede, epochs = 250, batch_size = 10)
```

```python
resultados = cross_val_score(estimator = rede_neural, X = X, y=y, cv=10, scoring='accuracy')
```

```python
resultados
```

```python
resultados.mean()
```

```python
resultados.std()
```

## Desafio - Tuning de parametros Rede Neural

### Estrutura

```python
from scikeras.wrappers import KerasClassifier
from tensorflow.keras import backend as k
from tensorflow.keras.models import Sequential
from sklearn.model_selection import GridSearchCV
```

```python
def criar_rede_tuning(optimizer, loss, kernel_initializer, activation, neurons):
    k.clear_session()
    rede_neural = Sequential([
        tf.keras.layers.InputLayer(shape = (4,)), 
        tf.keras.layers.Dense(units = neurons, activation = activation, kernel_initializer = kernel_initializer),
        tf.keras.layers.Dropout(rate = 0.2),# Aqui é a aplicação de um Dropout - não obrigatório para avaliar a rede_neural
        tf.keras.layers.Dense(units = neurons, activation = activation, kernel_initializer = kernel_initializer),
        tf.keras.layers.Dropout(rate = 0.2),
        tf.keras.layers.Dense(units = 3, activation = 'softmax')])
    otimizador = tf.keras.optimizers.Adam(learning_rate = 0.001, clipvalue = 0.5)
    rede_neural.compile(optimizer = optimizer, loss = loss, metrics = ['accuracy'])
    return rede_neural
```

```python
rede_neural = KerasClassifier(model = criar_rede_tuning)
```

```python
parametros = {
    'batch_size': [10,30],
    'epochs': [250,300],
    'model__optimizer': ['adam'],
    'model__loss':['categorical_crossentropy'],
    'model__kernel_initializer':['random_uniform','normal'],
    'model__activation': ['relu'],
    'model__neurons': [4,5]
}
parametros
```

```python
grid_search = GridSearchCV(estimator = rede_neural, param_grid = parametros, cv = 5)
```

```python
grid_search = grid_search.fit(X,y)
```

```python
print(grid_search)
```

```python
melhores_parametros = grid_search.best_params_
```

```python
print(melhores_parametros)
```

```python
melhor_precisao = grid_search.best_score_
print(melhor_precisao)
```

## Desafio - Classificar somente uma planta

```python
classificador = Sequential([
    tf.keras.layers.InputLayer(shape=(4,)),
    tf.keras.layers.Dense(units=5,activation='relu'),
    tf.keras.layers.Dropout(rate = 0.2),
    tf.keras.layers.Dense(units=5,activation='relu'),
    tf.keras.layers.Dropout(rate = 0.2),
    tf.keras.layers.Dense(units=3,activation='softmax'),
])
```

```python
classificador.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
classificador.fit(X,y, batch_size=10, epochs=300)
```

```python
X
```

```python
novo = np.array([[4.8, 3.4, 1.6, 0.2]])
novo.size, novo.shape
```

```python
previsao = classificador.predict(novo)
```

```python
previsao = previsao > 0.5
print(previsao)
```

```python
previsao_final = [np.argmax(t) for t in previsao] 
previsao_final
```
